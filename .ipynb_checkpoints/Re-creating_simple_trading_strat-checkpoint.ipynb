### This recreates an example from python for finance of a simple lt avg vs st avg trading signal strategy testing


```python


from datetime import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import pandas_datareader.data as web
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tickers = ["BIIB", "AMGN", "CELG", "GILD", "REGN"] #some nbi tickers
```

```python
stocks = {}

start = datetime(2015, 1, 1)

for tik in tickers:
    stocks[tik] = pd.DataFrame(web.DataReader(tik, "yahoo", start = start))
    
    
```

```python
stocks["BIIB"].head()
```

```python
fig = plt.figure(figsize = (15,10))

for tik in tickers:
    
    stocks[tik]['Close'].plot()

    
plt.legend(labels = tickers, loc = 0)
```

```python
#lets do a 20 day versus 10 day average trading strat like in the python finance book
for tik in tickers:
    stocks[tik]["10d"] = np.round(stocks[tik]['Close'].rolling(window = 10, center = False).mean(), 2)
    stocks[tik]["50d"] = np.round(stocks[tik]['Close'].rolling(window = 50, center = False).mean(), 2)
        
```

```python
stocks["BIIB"][['Close', '10d', '50d']].plot(grid=True, figsize=(15, 10))
```

```python
#set up difference between the two signals
for tik in tickers:
    stocks[tik]["10d-50d"] = stocks[tik]["10d"] - stocks[tik]["50d"]
```

The trading strtegy is as follows. For each name we will:

1. 10d breaches 20d + tolerance level ---> BUY (earn 1 * return)
2. 10d falls below 20d + tolerance level ---> cash (earn 0 return)
3. 10d crosses below 20d - tolerance level ---> SELL SHORT (earn -1 * return)

*ignore shorting / txn costs

```python
for tik in tickers:
    tolerance = stocks[tik]["Close"].rolling(window = 20).std()*2
    stocks[tik]["regime"] = np.where(stocks[tik]["10d-50d"] > tolerance , 1, 0)
    stocks[tik]["regime"] = np.where(stocks[tik]["10d-50d"] < -tolerance, -1, stocks[tik]["regime"])
```

```python
for tik in tickers:
    stocks[tik]["stock"] = np.log(stocks[tik]["Close"] / stocks[tik]["Close"].shift(1)) 
```

```python
for tik in tickers:
    stocks[tik]["strat"] = stocks[tik]["regime"].shift(1)*stocks[tik]["stock"]
```

```python
for tik in tickers:
    fig = plt.figure()
    stocks[tik][["strat", "stock"]].cumsum().apply(np.exp).plot(figsize = (15,10))
    plt.title(tik)
    plt.show()
    
```
